```python
import unsloth 
from unsloth import FastLanguageModel
from transformers import TextStreamer
from kaggle_secrets import UserSecretsClient
import os

user_secrets = UserSecretsClient()
os.environ['HF_TOKEN'] = user_secrets.get_secret("HF_TOKEN")
```

```python
MODEL = 'rtweera/2025-Jun-12_15-31-29_Qwen2.5-7B-Instruct-unsloth-bnb-4bit_LoRA'
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = MODEL,
    max_seq_length = 4096,
    dtype = None,
    load_in_4bit = True,
    token = os.getenv('HF_TOKEN')
)
FastLanguageModel.for_inference(model)

# Text streamer
streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
```

```python
def get_messages(user_prompt, system_prompt = None, tokenizer=tokenizer, model=model):
    """
    Prepare messages for the model.
    """
    messages = []
    if system_prompt:
        messages.append({"role": "system", "content": system_prompt})
    messages.append({"role": "user", "content": user_prompt})
    tokenized_msg = tokenizer.apply_chat_template(messages, add_generation_prompt=True, tokenize=True, return_tensors="pt").to(model.device)
    return tokenized_msg
```

```python
def stream_sample(sample_question, system_prompt = None):
    """
    Stream the model's response to a sample question.
    """
    tokenized_msg = get_messages(sample_question, system_prompt)
    
    model.generate(
        input_ids=tokenized_msg,
        streamer=streamer,
        max_new_tokens=2048,
        use_cache=True,
        temperature=0.1,
        min_p=0.1,
    )
```

```python
while (x:=input('You >>  ')).strip()!='/exit':
    print('Assistant >> ', end='')
    stream_sample(x)
```
